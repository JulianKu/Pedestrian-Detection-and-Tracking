```
import os
import shutil as sh
```

Define checkpoint that should be stored externally

```
ckpt = 26500 # insert most recent checkpoint number here
```

```
pathckpt = '/content/ckpt_{}'.format(ckpt)
if not os.path.exists(pathckpt):
  os.makedirs(pathckpt)
sh.copyfile('/content/darkflow/ckpt/checkpoint', os.path.join(pathckpt, 'checkpoint'))

darkckpt = '/content/darkflow/ckpt'
ckptfiles = os.listdir(darkckpt)
for file in ckptfiles:
  if str(ckpt) in file:
    sh.copyfile(os.path.join(darkckpt,file), os.path.join(pathckpt, file))
```

```
# Install the PyDrive wrapper & import libraries.
# This only needs to be done once in a notebook.
!pip install -U -q PyDrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once in a notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

Upload file to Google Drive

```
# Create & upload a file.
uploaded = drive.CreateFile({'title': 'ckpt{}.zip'.format(ckpt)})
uploaded.SetContentFile('/content/ckpt{}.zip'.format(ckpt))
uploaded.Upload()
print('Uploaded file with ID {}'.format(uploaded.get('id')))
```
